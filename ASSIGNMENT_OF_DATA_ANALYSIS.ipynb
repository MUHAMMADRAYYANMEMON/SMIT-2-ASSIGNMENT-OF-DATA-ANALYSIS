# **Data Visualization**

# **1.1 Two Types of Data**

```python
# Plot a correlation heatmap to visualize relationships between numeric variables
sns.heatmap(df[num_features].corr(), annot=True, cmap='coolwarm')
```

```python
#Numerical + Numerical
plt.scatter(df['Profit'],df['Discount'],color='blue',marker='+')
plt.title('profit with discount analysis')
plt.xlabel('Profit')
plt.ylabel('Discount')
```

```python
#category + category
sns.countplot(x='Segment', hue='Category', data=df)
plt.title('Countplot with Hue')
plt.show()
```

```python
#Categorical vs Numerical

sns.barplot(x='Category', y='Sales', data=df)
plt.title('Bar Plot of category_column vs numeric_column')
plt.show()
```

```python
#Insights
top_cities = df['City'].value_counts().head(10).index
sns.boxplot(x='City', y='Sales', data=df[df['City'].isin(top_cities)])
plt.xticks(rotation=45)
plt.show()



```
